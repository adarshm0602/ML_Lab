```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 100

data = pd.DataFrame({
    'X' : np.random.normal(0, 1, n),
    'Y' : np.random.normal(0, 1, n),
    'Z' : np.random.normal(0, 1, n),
    'Category' : np.random.choice(['A', 'B', 'C'], n)
})
```

```python
def scatter_plot():
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=data, x='X', y='Y', hue='Category')
    plt.title("2D scatter plot")
    plt.show()
```

```python
scatter_plot()
```

```python
def box_plot():
    plt.figure(figsize=(6,4))
    sns.boxplot(data=data, x='Category',y='Z')
    plt.title("Box Plot of Z by Category")
    plt.show()
```

```python
box_plot()
```

```python
def heat_map():
    correlations = data[['X','Y','Z']].corr()
    plt.figure(figsize=(5,4))
    sns.heatmap(correlations,annot=True,cmap='coolwarm')
    plt.title("Heatmap of correaltions matrix")
    plt.show()
```

```python
heat_map()
```

```python
def contour_plot():
 x = np.linspace(-3, 3, 100)
 y = np.linspace(-3, 3, 100)
 X, Y = np.meshgrid(x, y)
 Z = np.sin(X**2 + Y**2)
 plt.figure(figsize=(6, 5))
 cp = plt.contourf(X, Y, Z, cmap='viridis')
 plt.colorbar(cp)
 plt.title("Contour Plot of sin(X + Y)")
 plt.xlabel("X")
 plt.ylabel("Y")
 plt.show() 
```

```python
contour_plot()
```

```python
def surface_plot():
 fig = plt.figure(figsize=(8, 6))
 ax = fig.add_subplot(111, projection='3d')
 x = np.linspace(-3, 3, 100)
 y = np.linspace(-3, 3, 100)
 X, Y = np.meshgrid(x, y)
 Z = np.sin(np.sqrt(X**2 + Y**2))
 surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
 fig.colorbar(surf)
 ax.set_title("3D Surface Plot")
 plt.show() 
```

```python
surface_plot()
```
